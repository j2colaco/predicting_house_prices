```python
# lets read in the data
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
from math import log
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectPercentile
%matplotlib inline

filepath = '//Users//joanne//Desktop//predicting_house_prices//data//'
data = pd.read_csv(filepath + 'train.csv')
test_data = pd.read_csv(filepath + 'test.csv')
print(data.shape)
```

```python
# Select all the numeric variables and drop columns that have null values
house_features = (data
                 .select_dtypes('number')
                 .drop(columns=['Id', 'SalePrice', 'LotFrontage', 'MasVnrArea', 'GarageYrBlt']))

house_y = data['SalePrice']
```

```python
# look at which columns contain null values
house_features.isnull().sum()

# Split the data into training and test sets
train_X, test_X, train_y, test_y = train_test_split(house_features, house_y, test_size=0.2, random_state=34234)
```

```python
# Multi-Linear Regression model using all numeric features
model_all_features = sm.OLS(train_y, train_X).fit()
model_all_features.summary()
```

```python
# Let's get the train and test RMSE to compare
train_predictions = model_all_features.predict(train_X)
print('The training RMSE is ', sqrt(mean_squared_error(train_y, train_predictions)), '!')

test_predictions = model_all_features.predict(test_X)
print('The testing RMSE is ', sqrt(mean_squared_error(test_y, test_predictions)), '!')
```

```python
# View the residual plot
model_all_features_residuals = (test_y-test_predictions)
# model_all_features_residuals


# model_all_features_residuals_plt = data.plot(kind="scatter", x='LotArea', y='SalePrice')
# plt.show()
```

```python
# fitted values (need a constant term for intercept)
model_all_features_fitted = model_all_features.fittedvalues
print(model_all_features_fitted.head())
# model residuals
model_all_features_residuals = model_all_features.resid
print(model_all_features_residuals.head())
```

```python
# Let's submit the predictions of this model and record the score
# Select all the numeric variables from test data and drop columns that have null values
test_features = (test_data
                 .select_dtypes('number')
                 .drop(columns=['Id', 'LotFrontage', 'MasVnrArea', 'GarageYrBlt']))

submit_predictions = model_all_features.predict(test_features)
# print(submit_predictions)
```

```python
submit_predictions.isnull().sum()
# Fill in nulls with the mean
submit_predictions = pd.DataFrame(submit_predictions)
submit_predictions.columns = ['SalePrice']
submit_predictions = submit_predictions.SalePrice.fillna(submit_predictions.SalePrice.mean())
# Put in the Id column required for submission
submit_predictions = pd.concat([test_data['Id'], submit_predictions], axis=1)
```

```python
# Submit Model with all numeric feature predictions (Score: 0.59828)
submit_filepath = '//Users//joanne//Desktop//predicting_house_prices//'
submit_predictions.to_csv(submit_filepath + 'model_all_features_predictions.csv', 
                          index=False, header=True)
```

```python
# Scale all numeric features with standard scaler. Mean of zero and unit variance of 1
scale_all_features = StandardScaler().fit_transform(house_features)
# scale_all_features.mean_
# scale_all_features.scale_


# Split to test and training set
scale_train_X, scale_test_X, scale_train_y, scale_test_y = train_test_split(
    scale_all_features, house_y, test_size=0.2, random_state=34234)


# Building a model with scaled features
model_scale_all_features = sm.OLS(scale_train_y, scale_train_X).fit()
model_scale_all_features.summary()

# Got a really low R_squared, not sure why
```

```python
# Scale all numeric features with MinMaxScaler. Scales based on the range
scale_all_features = MinMaxScaler().fit_transform(house_features)

# Split to test and training set
scale_train_X, scale_test_X, scale_train_y, scale_test_y = train_test_split(
    scale_all_features, house_y, test_size=0.2, random_state=34234)

# Building a model with scaled features
model_scale_all_features = sm.OLS(scale_train_y, scale_train_X).fit()
model_scale_all_features.summary()

# Got a slightly lower R_squared than model_all_features
```

```python
# replaces all the null values in the test_data with mean column values
for col in test_features.columns:
    if test_features[col].isnull().sum() > 0:
        test_features[col] = test_features[col].fillna(test_features[col].mean())

# test_features.isnull().sum()

# Submit predictions for model_scale_all_features
# scale test data
scale_test_features = MinMaxScaler().fit_transform(test_features)

model_scale_all_features_pred = model_scale_all_features.predict(scale_test_features)

model_scale_all_features_pred = pd.DataFrame(model_scale_all_features_pred)
print(model_scale_all_features_pred.isnull().sum())
# Fill in nulls with the mean
model_scale_all_features_pred.columns = ['SalePrice']
# Put in the Id column required for submission
model_scale_all_features_pred = pd.concat([test_data['Id'], model_scale_all_features_pred], axis=1)
```

```python
# Export submit csv (Score: 0.70633)
model_scale_all_features_pred.to_csv(submit_filepath + 'scale_model_all_features_pred.csv', 
                          index=False, header=True)

# Why did I get a worse score after scaling the entire dataset?
# Possible because all features are now on the same scale, they will have equal importance on the SalePrice so
# features with a larger range have larger weightage but are also more important in to determining SalesPrice 
# accurately
```

```python
# Mean centering all the numeric features (Different way of standardizing features)
center_all_features = preprocessing.scale(house_features, with_mean=True, with_std=False)

center_train_X, center_test_X, center_train_y, center_test_y = train_test_split(
    center_all_features, house_y, test_size=0.2, random_state=43234)

# Build a model
model_center_all_features = sm.OLS(center_train_y, center_train_X).fit()
model_center_all_features.summary()

# Again got a really low R squared. It is because there is now a possiblilty of predicting
# a negative sales price which is not possible.
# What we have to do is center the sales price data first and predict the centering. Then
# we have to work backwards to get the original sales price data. Let's try this!
```

```python
house_y_mean = house_y.mean()
house_y_mean

center_house_y = preprocessing.scale(house_y, with_mean=True, with_std=False)

center_train_X, center_test_X, center_train_y, center_test_y = train_test_split(
    center_all_features, center_house_y, test_size=0.2, random_state=43344)

# Build a model
model_center_all_features = sm.OLS(center_train_y, center_train_X).fit()
model_center_all_features.summary()

# As hypothesized, got a much high r squared than before
```

```python
# Submit center model predictions to test the score
test_features.isnull().sum()

# Submit predictions for model_scale_all_features
# scale test data
center_test_features =  preprocessing.scale(test_features, with_mean=True, with_std=False)

model_center_all_features_pred = model_center_all_features.predict(center_test_features)

model_center_all_features_pred = pd.DataFrame(model_center_all_features_pred)
print(model_center_all_features_pred.isnull().sum())
model_center_all_features_pred.columns = ['SalePrice']
# Put in the Id column required for submission
model_center_all_features_pred = pd.concat([test_data['Id'], model_center_all_features_pred], axis=1)
```

```python
# Add the mean to the centered sales price
model_center_all_features_pred['SalePrice'] = model_center_all_features_pred['SalePrice'] + house_y_mean
```

```python
# Export submit csv (Score: 0.45897)
model_center_all_features_pred.to_csv(submit_filepath + 'center_model_all_features_pred.csv', 
                          index=False, header=True)
```

```python
# Let's explore feature selection for only numeric features
# How many numeric features do we have
print('There are ', house_features.shape[1], 'numeric features in the dataset.')

# Remove numeric features that do not meet a variance threshold of 80%
var_thres = VarianceThreshold(threshold=(.8 * (1 - .8)))
# var_thres = VarianceThreshold(threshold=(.8))
var_thres.fit(house_features)
var_threshold_features_cols = house_features[house_features.columns[var_thres.get_support(indices=True)]].columns

# Subset data to just the features that meet the variance threshold
var_threshold_features = house_features[var_threshold_features_cols]


print('There are ', var_threshold_features.shape[1], 'numeric features',
      'in the dataset that meet variance threshold criteria.')
```

```python
# Build a model using selected features without scaling data
var_thres_train_X, var_thres_test_X, train_y, test_y = train_test_split(
    var_threshold_features, house_y, test_size=0.2, random_state=43344)

# Build a model
model_var_thres_features = sm.OLS(train_y, var_thres_train_X).fit()
model_var_thres_features.summary()

# Highest R squared I have ever gotten. I hypothesize that my submission will not be higher
# as it may overfit the training data
```

```python
# Submit predictions for model_var_thres_features
# Not scaling the test_data, just need to remove the features that
# did not meet the variance threshold
var_threshold_test_features = test_features[var_threshold_features_cols]

model_var_thres_features_pred = model_var_thres_features.predict(var_threshold_test_features)

# Naming the dataframe columns and concat id columns
print(model_var_thres_features_pred.isnull().sum())
model_var_thres_features_pred = pd.DataFrame(model_var_thres_features_pred)
model_var_thres_features_pred.columns = ['SalePrice']
# Put in the Id column required for submission
model_var_thres_features_pred = pd.concat([test_data['Id'], model_var_thres_features_pred], axis=1)
```

```python
# Export submit csv (Score: 0.47817 [0.8*(1-0.8)], Score: 0.92004 [80%])
model_var_thres_features_pred.to_csv(submit_filepath + 'var_thres_(80)_features_pred.csv', 
                          index=False, header=True)
```

```python
# I realized that I have been using variance threshold all wrong
# I should have first normalized the data and then applied variance threshold
# Thus, I will not be using variance threshold anymore.
# Based on the previous vairance threshold, I will center the features and predict
# the test data



# # Change center_all_features from numpy to dataframe and add column names
center_all_features_df = pd.DataFrame(center_all_features)
center_all_features_df.columns = house_features.columns

var_thres_center_all_features = center_all_features_df[var_threshold_features_cols]

center_var_thres_train_X, center_var_thres_test_X, center_train_y, center_test_y = train_test_split(
    var_thres_center_all_features, center_house_y, test_size=0.2, random_state=43344)

# Build a model
model_center_var_thres_features = sm.OLS(center_train_y, center_var_thres_train_X).fit()
model_center_var_thres_features.summary()
```

```python
# Submit center model predictions to test the score

# Submit predictions for model_scale_all_features
# scale test data
center_test_features =  preprocessing.scale(test_features, with_mean=True, with_std=False)
center_test_features_df = pd.DataFrame(center_test_features)
center_test_features_df.columns = house_features.columns

var_thres_center_test_features = center_test_features_df[var_threshold_features_cols]

model_var_thres_center_test_features_pred = model_center_var_thres_features.predict(var_thres_center_test_features)

model_var_thres_center_test_features_pred = pd.DataFrame(model_var_thres_center_test_features_pred)
print(model_var_thres_center_test_features_pred.isnull().sum())
model_var_thres_center_test_features_pred.columns = ['SalePrice']
# Put in the Id column required for submission
model_var_thres_center_test_features_pred = pd.concat([test_data['Id'], model_var_thres_center_test_features_pred], axis=1)
```

```python
# Add the mean to the centered sales price
model_var_thres_center_test_features_pred['SalePrice'] = model_var_thres_center_test_features_pred['SalePrice'] + house_y_mean
```

```python
# Export submit csv (Score: 0.46229)
model_var_thres_center_test_features_pred.to_csv(submit_filepath + 'center_model_var_thres_features_pred.csv', 
                          index=False, header=True)
```

```python
# NEXT STEP Feature selection based on percentile (Selects the best percentile!)
percentile_select = SelectPercentile(score_func=chi2, percentile=20)
percentile_fit = percentile_select.fit(house_features, house_y)
percentile_features_cols = house_features[house_features.columns[percentile_fit.get_support(indices=True)]].columns

print(percentile_features_cols)

percentile_features = house_features[percentile_features_cols]

percentile_features_train_X, percentile_features_test_X, train_y, test_y = train_test_split(
    percentile_features, house_y, test_size=0.2, random_state=43344)

# Build the model
model_percentile_features = sm.OLS(train_y, percentile_features_train_X).fit()
model_percentile_features.summary()
```

```python
# Submit predictions
# Not scaling the test_data, just need to remove the features that
# did not meet the percentile
percentile_test_features = test_features[percentile_features_cols]

model_percentile_features_pred = model_percentile_features.predict(percentile_test_features)

# Naming the dataframe columns and concat id columns
print(model_percentile_features_pred.isnull().sum())
model_percentile_features_pred = pd.DataFrame(model_percentile_features_pred)
model_percentile_features_pred.columns = ['SalePrice']
# Put in the Id column required for submission
model_percentile_features_pred = pd.concat([test_data['Id'], model_percentile_features_pred], axis=1)
```

```python
# Export submit csv (Score: 0.72435 [10])
model_percentile_features_pred.to_csv(submit_filepath + 'model_percentile20_features_pred.csv', 
                          index=False, header=True)
```
